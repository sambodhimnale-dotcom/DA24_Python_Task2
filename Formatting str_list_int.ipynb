## **String Methods (str)**

##### *1. upper() / lower()*

```python
"This is a string".upper()
```

```python
"This is a string".lower()
```

#### *2. strip()*

```python
"   This is a string     ".strip()
```

#### *3. split(sep)*

```python
"This is a string1, this is string2".split(",")
```

#### *4. replace()*

```python
"This is a string".replace("string", "sentence")
```

#### *5. find()*

```python
"This is a sentence".find("ence")
```

#### *6. startswith(val)*

```python
"This is a sentence".startswith("This")
```

```python
"This is a sentence".startswith("That")
```

## **List Methods (list)**

#### *1. append()*

```python
list1 = [135, 246, 357]
list1.append(678)
print(list1)
```

#### *2. extend()*

```python
list1 = [135, 246, 357]
list1.extend([5,8])
print(list1)
```

#### *3. Insert()*

```python
list1 = [135, 246, 357]
list1.insert(1, 4)
print(list1)
```

#### *4. pop()*

```python
list1 =[135, 246, 357, 5, 8, 6]
list1.pop()
print(list1)
```

#### *5. remove()*

```python
list1 =[135, 246, 357, 5, 8, 6]
list1.remove(5)
print(list1)
```

#### *6. sort()*

```python
list1 =[135, 246, 357, 5, 8, 6]
list1.sort()
print(list1)
```

## **Integer Methods (int)**

#### *1. bit_length()*

```python
n = 3589
print(n.bit_length())
```

#### *2. numerator*

```python
n = 1087
print(n.numerator)
```

#### *3. Denominator*

```python
n = 1087
print(n.denominator)
```

#### *4. Different Bases*

```python
x = 2026
print(bin(x))
print(hex(x))
print(oct(x))
```
